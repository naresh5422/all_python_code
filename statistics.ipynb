##### Central Limit Theorem (CLT)

The CLT states that the distribution of sample means approaches a normal distribution as the sample size increases, regardless of the population distribution, given that the sample size is sufficiently large.

```python
# Generates a non normal distribution(Exponantial)
population = np.random.exponential(scale=2, size = 10000)
population
```

```python
plt.plot(population)
```

```python
# Draw multiple samples and calculate means
sample_means = [np.mean(np.random.choice(population, 30)) for _ in range(1000)]
len(sample_means)
```

```python
plt.figure(figsize = (6,4))
plt.hist(sample_means, bins = 30, density=True, alpha = 0.5, color='g')
plt.title('Sample Means Distribution (Central Limit Theorem)')
plt.show()
```

##### Variance
Variance is measures how much the data points in dataset far from the mean



```python
data = [10, 12, 23, 23, 16, 23, 21, 16]
variance = np.var(data)
print(f"Variance: {variance}")
```

```python
var = 1/len(data)*sum((data - np.mean(data))**2)
print(f"Variance: {var}")
```

##### Correlation

It quantifies the strength of the linear relationship between two variables,  pearson correlation coefficient most commonly used.

```python
x = np.random.rand(50)
y = 2*x+np.random.normal(0, 0.1, 50)
```

```python
corr = np.corrcoef(x,y)
corr[1,0]
```

##### Covariance
It is a direction of the linear relationship between two variables

```python
cov_matrix = np.cov(x,y)
print(f"Covariance Matrix:\n {cov_matrix}")
```

#### Standard Deviation
 it is the square root of variance, measuring the amount of variation or dispersion

```python
var = sum((data-np.mean(data))**2)/len(data)
std = np.sqrt(var)
print(f"Using Formula Standard Deviation: {std}")
```
